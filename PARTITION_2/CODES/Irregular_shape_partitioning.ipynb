```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 7))

# Add the large grey rectangle (irregular enclosing partition)
outer_rect = patches.Rectangle((5, 2), 80, 23, linewidth=2,
                               edgecolor='black', facecolor='lightgrey', zorder=1)
ax.add_patch(outer_rect)

# Define the dotted query rectangles (x, y, width, height)
queries = {
    "q1": {"xy": (48, 15), "width": 6, "height": 3, "color": "red"},
    "q2": {"xy": (50, 18), "width": 10, "height": 3, "color": "blue"},
    "q3": {"xy": (10, 14), "width": 30, "height": 4, "color": "green"},
    "q4": {"xy": (60, 3), "width": 20, "height": 6, "color": "purple"},
}

# Draw each query as a dotted rectangle
for qname, q in queries.items():
    rect = patches.Rectangle(q["xy"], q["width"], q["height"],
                             linewidth=2, edgecolor=q["color"],
                             facecolor='none', linestyle='dotted', zorder=2)
    ax.add_patch(rect)
    ax.text(q["xy"][0] + q["width"]/2, q["xy"][1] + q["height"] + 0.5,
            qname.upper(), color=q["color"], ha='center', va='bottom', fontsize=9, weight='bold')

# Plot formatting
ax.set_xlim(0, 90)
ax.set_ylim(0, 25)
ax.set_title("Irregular Partition: Grey Region Surrounding All Queries")
ax.set_xlabel("c_acctbal (normalized 100–9000 → 0–90)")
ax.set_ylabel("c_nationkey (0–24)")
plt.grid(False)

# Optional: save image
# plt.savefig("Irregular_Grey_Region_With_Queries.png", bbox_inches="tight")
plt.tight_layout()
plt.show()
```
