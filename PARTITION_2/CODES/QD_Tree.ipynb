```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Define tree structure with fixed binary layout
tree = nx.DiGraph()

# Tree structure based on your manual QD-tree
tree.add_edges_from([
    ("Root", "Root.L"), ("Root", "Root.R"),
    ("Root.L", "P1"), ("Root.L", "P2"),
    ("Root.R", "Root.R.L"), ("Root.R", "P5"),
    ("Root.R.L", "P3"), ("Root.R.L", "P4")
])
```

```python
# Labels for all nodes
labels = {
    "Root": "Split: c_acctbal ≤ 4600",
    "Root.L": "Split: c_nationkey ≤ 12",
    "Root.R": "Split: c_nationkey ≤ 12",
    "Root.R.L": "Split: c_acctbal ≤ 6800",
    "P1": "P1\n140MB",
    "P2": "P2\n138MB",
    "P3": "P3\n119MB",
    "P4": "P4\n118MB",
    "P5": "P5\n136MB"
}
```

```python
# Manual positions to simulate top-down binary tree
pos = {
    "Root": (0, 5),
    "Root.L": (-3, 4),
    "Root.R": (3, 4),
    "P1": (-4, 3),
    "P2": (-2, 3),
    "Root.R.L": (2, 3),
    "P5": (4, 3),
    "P3": (1.5, 2),
    "P4": (2.5, 2)
}
```

```python
# Draw the tree
plt.figure(figsize=(10, 6))
nx.draw(tree, pos, with_labels=False, arrows=True, node_size=2500, node_color='lightblue', edge_color='black')
nx.draw_networkx_labels(tree, pos, labels=labels, font_size=10)
plt.title("Proper QD-Tree with 5 Leaf Nodes and Partition Sizes", fontsize=13)
plt.axis('off')
plt.tight_layout()
plt.show()
```
