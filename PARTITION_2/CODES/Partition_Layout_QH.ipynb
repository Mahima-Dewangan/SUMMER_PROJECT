```python
# Re-import necessary libraries due to kernel reset
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Re-define partition layout
partitions = [
    {'id': 'P1', 'x': 0, 'y': 12, 'width': 46, 'height': 13, 'color': 'gray', 'size': 140},
    {'id': 'P2', 'x': 0, 'y': 0,  'width': 46, 'height': 12, 'color': 'lightgray', 'size': 138},
    {'id': 'P3', 'x': 46, 'y': 12, 'width': 22, 'height': 13, 'color': 'lightcoral', 'size': 119},
    {'id': 'P4', 'x': 68, 'y': 12, 'width': 22, 'height': 13, 'color': 'wheat', 'size': 118},
    {'id': 'P5', 'x': 46, 'y': 0,  'width': 44, 'height': 12, 'color': 'paleturquoise', 'size': 136},
]

# Define 4 queries: 2 in P3, 1 in P1, 1 in P5
queries = [
    {'id': 'q1', 'x': 47, 'y': 14, 'width': 5, 'height': 3, 'color': 'red'},
    {'id': 'q2', 'x': 50, 'y': 17, 'width': 10, 'height': 3, 'color': 'blue'},
    {'id': 'q3', 'x': 10, 'y': 14, 'width': 30, 'height': 5, 'color': 'green'},
    {'id': 'q4', 'x': 60, 'y': 3, 'width': 20, 'height': 6, 'color': 'purple'},
]

# Plot partitions and queries
fig, ax = plt.subplots(figsize=(8, 6))

# Draw partitions
for p in partitions:
    rect = patches.Rectangle((p['x'], p['y']), p['width'], p['height'],
                             linewidth=2, edgecolor='black', facecolor=p['color'], alpha=0.4)
    ax.add_patch(rect)
    ax.text(p['x'] + p['width']/2, p['y'] + p['height']/2,
            f"{p['id']}\n{p['size']}MB", fontsize=9, ha='center', va='center', color='black')

# Draw query boxes
for q in queries:
    rect = patches.Rectangle((q['x'], q['y']), q['width'], q['height'],
                             linewidth=2, edgecolor=q['color'], facecolor='none', linestyle='--', label=q['id'])
    ax.add_patch(rect)

# Axis labels and legend
ax.set_xlim(0, 90)
ax.set_ylim(0, 25)
ax.set_xlabel("c_acctbal (normalized 100–9000 → 0–90)")
ax.set_ylabel("c_nationkey (0–24)")
ax.set_title("QD-Tree Layout with 4 Range Queries")

# Unique legend
handles = [patches.Patch(edgecolor=q['color'], facecolor='none', linestyle='--', label=q['id']) for q in queries]
plt.legend(handles=handles, title="Queries", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
