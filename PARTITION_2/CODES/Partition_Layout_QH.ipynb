```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Partition layout (based on actual c_acctbal and c_nationkey ranges)
partitions = [
    {'id': 'P1', 'x': 100, 'y': 0, 'width': 4600 - 100, 'height': 13, 'color': 'skyblue', 'size': 140},
    {'id': 'P2', 'x': 100, 'y': 13, 'width': 4600 - 100, 'height': 12, 'color': 'lightgreen', 'size': 138},
    {'id': 'P3', 'x': 4601, 'y': 0, 'width': 6800 - 4601, 'height': 13, 'color': 'salmon', 'size': 150},
    {'id': 'P4', 'x': 6801, 'y': 0, 'width': 9000 - 6801, 'height': 13, 'color': 'violet', 'size': 128},
    {'id': 'P5', 'x': 4601, 'y': 13, 'width': 9000 - 4601, 'height': 12, 'color': 'gold', 'size': 136}
]

# Define 4 queries (coordinates in same scale as acctbal and nationkey)
queries = [
    {'id': 'q1', 'x': 5000, 'y': 14, 'width': 500, 'height': 3, 'color': 'red'},
    {'id': 'q2', 'x': 5500, 'y': 17, 'width': 400, 'height': 3, 'color': 'blue'},
    {'id': 'q3', 'x': 1000, 'y': 14, 'width': 3000, 'height': 5, 'color': 'green'},
    {'id': 'q4', 'x': 6000, 'y': 3, 'width': 500, 'height': 6, 'color': 'purple'},
]

# Plot partitions and queries
fig, ax = plt.subplots(figsize=(10, 6))

# Draw partitions
for p in partitions:
    rect = patches.Rectangle((p['x'], p['y']), p['width'], p['height'],
                             linewidth=2, edgecolor='black', facecolor=p['color'], alpha=0.4)
    ax.add_patch(rect)
    ax.text(p['x'] + p['width']/2, p['y'] + p['height']/2,
            f"{p['id']}\n{p['size']}MB", fontsize=9, ha='center', va='center', color='black', weight='bold')

# Draw query boxes
for q in queries:
    rect = patches.Rectangle((q['x'], q['y']), q['width'], q['height'],
                             linewidth=2, edgecolor=q['color'], facecolor='none', linestyle='--', label=q['id'])
    ax.add_patch(rect)

# Axis labels and limits
ax.set_xlim(100, 9100)
ax.set_ylim(0, 25)
ax.set_xlabel("c_acctbal (1009000)")
ax.set_ylabel("c_nationkey (024)")
ax.set_title("QD-Tree Partition Layout with 4 Range Queries")

# Unique legend
handles = [patches.Patch(edgecolor=q['color'], facecolor='none', linestyle='--', label=q['id']) for q in queries]
plt.legend(handles=handles, title="Queries", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
