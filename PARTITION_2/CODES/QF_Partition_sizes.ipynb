```python
import matplotlib.pyplot as plt
import pandas as pd

# Final partition sizes
partition_sizes = {
    "P1": 140,
    "P2": 138,
    "P3": 150,
    "P4": 128,
    "P5": 136
}

# Query to partition overlap mapping based on your image and instructions
query_partition_map = {
    "Q1 (Red)": ["P5"],
    "Q2 (Blue)": ["P1", "P2", "P3", "P5"],
    "Q3 (Green)": ["P1", "P2", "P3", "P4", "P5"],
    "Q4 (Purple)": ["P1", "P3", "P4"]
}

# Build data for DataFrame
query_data = []
for query, parts in query_partition_map.items():
    total_size = sum(partition_sizes[p] for p in parts)
    query_data.append({
        "Query": query,
        "Partitions Touched": ", ".join(parts),
        "Total Size (MB)": total_size
    })

# Create DataFrame
df = pd.DataFrame(query_data)

# Plot the table
fig, ax = plt.subplots(figsize=(9, 2.2))
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)

# Save as PNG
# png_path = "/mnt/data/final_verified_query_partition_table.png"
# plt.savefig(png_path, bbox_inches="tight")
# plt.close()

# print("✅ Table generated and saved as PNG:")
# print(png_path)
```
