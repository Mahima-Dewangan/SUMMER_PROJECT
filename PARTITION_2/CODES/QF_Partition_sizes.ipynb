```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Data: Partition sizes touched by each query
data = {
    "Query": ["Q1 (Red)", "Q2 (Green)", "Q3 (Blue)", "Q4 (Purple)"],
    "Partitions Touched": [
        "P3",
        "P1, P2, P3, P5",
        "All 5 partitions",
        "P2, P5"
    ],
    "Total Size (MB)": [
        110,
        125 + 130 + 110 + 120,
        125 + 130 + 110 + 115 + 120,
        130 + 120
    ]
}

df = pd.DataFrame(data)

# Plot table
fig, ax = plt.subplots(figsize=(8, 2))
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)


```
